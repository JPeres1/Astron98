```python
#Preamble- Import statements
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x,y = np.meshgrid(np.linspace(-1,1,100), np.linspace(-2,2,100))
f = x + y
                  
#figure sandwhich
plt.figure()
plt.title('x+y')
plt.xlabel('x')
plt.ylabel('y')
plt.imshow(f, cmap = 'inferno', origin = 'lower')
plt.colorbar()
plt.show()                  
```

```python
x,y = np.meshgrid(np.linspace(-5,5,100), np.linspace(5,5,100))
f = np.sqrt(x**2 + y**2)
                  
#figure sandwhich
plt.figure()
plt.title('x+y')
plt.xlabel('x')
plt.ylabel('y')
plt.imshow(f, cmap = 'inferno', origin = 'lower')
plt.colorbar()
plt.show()                  
```
