## Problem 2: van der Waals Gas

A Van der Waals gas is a model of gas particles that takes into account the interaction of the particles and the volume the particles take up. Like the ideal gas law, it relates pressure, volume, amount of gas and temperature. The equation is <br>
$\left(P +a\frac{n^2}{V^2}\right)(V-nb) = nRT \quad (1)$ <br>
Where $P$ is pressure, $V$ is volume, $T$ is temperature, $n$ is number of moles, $R$ is the ideal gas constant and $a,b$ are parameters. The value of $R$ is $0.083144 \frac{barL}{molK}$

Make a colormap plot of the temperature of one mole acetone for given volume and pressure. The pressure should vary between 1 and 10 bar and the volume should vary between 10 and 30 liters. For acetone $a = 16.02 \frac{barL^2}{mol^2}$ and $b = 0.1124 \frac{L}{mol}$. Make sure your axes labels have the correct values. Also make use of `vmin` and `vmax` to show detail in your color map and add a color bar to the map



```python
#I started with the equation first, but had to set it before the equation because I was receiving an error about P and V not being defined
#make pressure and volume ranges and create grid of values
pressure_range = np.linspace(0, 10, 100)  # pressure from 1 to 10 bar
volume_range = np.linspace(10, 30, 100)  # volume from 10 to 30 liters
P, V = np.meshgrid(pressure_range, volume_range)

#label constants
R=0.083144 #barL/molK
a=16.02 #barL**2/mol**2
b= 0.1124 #L/mol
n=1

#write equation
T=(P + a*(n**2)/(V**2))*(V-n*b)/n*R





# Create a colormap plot
plt.figure(figsize=(10, 8))
cmap = plt.get_cmap('viridis')
plot = plt.contourf(P, V, T, cmap=cmap, levels=100)

plt.xlabel('Pressure')
plt.ylabel('Volume')
plt.title('Temperature')
cbar = plt.colorbar(plot)
cbar.set_label('Temperature')

# Set the range for the color map
cbar.set_clim(vmin=np.min(T), vmax=np.max(T))

# Show the plot
plt.show()
```

## Problem 4: Monte Carlo $\pi$ ##

Did you know that we can calculate the value of $\pi$ by generating random numbers? We first generate a whole bunch of random points in a 1x1 square with the corner at the origin. Then we see how many of the points are within a distance 1 of the origin. There points all fall in a quarter of a circle centered on the origin. Then using the relation <br>
$\frac{N_{in}}{N} \approx \frac{A_{qc}}{A}$ <br>
we can calculate $\pi$ from the ratio of the areas of a quarter circle and the square.

Calculate the value of $\pi$ using this procedure for N = 1e3, 1e4, 1e5, 1e6, and print your results. Also create a plot of these points for N = 1e4. The points within the quarter circle should all be one color and the points outside should also be a different color. Also please make sure your plot is a square and put the value of pi from that run on your plot as text.



```python
def estimate_pi(N):
    # Generate random points in the 1x1 square
    points = np.random.rand(int(N), 2)

    # Check if points fall within the quarter circle
    inside_circle = points[:, 0]**2 + points[:, 1]**2 <= 1

    # Calculate the ratio of points inside the quarter circle to the total points
    ratio = np.sum(inside_circle) / N

    # Estimate  using the ratio
    pi_estimate = 4 * ratio

    return pi_estimate, points, inside_circle

# Values of N to test
N_values = [1e3, 1e4, 1e5, 1e6]

# Loop over N values, estimate , and print the result
for N in N_values:
    pi_estimate, _, _ = estimate_pi(N)
    print(f'For N = {N}: Estimated  = {pi_estimate}')

# Create a plot for N = 1e4
pi_estimate_1e4, points, inside_circle = estimate_pi(1e4)

# Plot points within and outside the quarter circle
plt.figure(figsize=(6, 6))
plt.scatter(points[:, 0], points[:, 1], c=inside_circle, cmap='viridis', s=5)

# Draw the quarter circle
theta = np.linspace(0, np.pi/2, 100)
x_circle = np.cos(theta)
y_circle = np.sin(theta)
plt.plot(x_circle, y_circle, color='red', linestyle='--', label='Quarter Circle')

# Set plot limits and aspect ratio
plt.xlim(0, 1)
plt.ylim(0, 1)
plt.gca().set_aspect('equal', adjustable='box')

# Annotate the plot with the estimated  value
plt.text(0.1, 0.8, f'Estimated  = {pi_estimate_1e4:.5f}', fontsize=10, bbox=dict(facecolor='white', alpha=0.5))

# Set axis labels and title
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Monte Carlo Estimation of ')

# Show the plot
plt.show()
```
